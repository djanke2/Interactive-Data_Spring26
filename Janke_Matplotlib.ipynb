```python
from matplotlib import pyplot as plt
import pandas as pd
```

I would like to learn how to make pie charts.  I believe it would be a good representation for what percentage each different category of revenue contributed to the overall total Revenue for a company.

```python
cces = pd.read_csv('cces_example_data.csv')
cces.head()
```

```python
fig1, ax1 = plt.subplots()
plt.show()
```

```python
fig2, axs2 = plt.subplots(1, 2)
axs2[0].set_xlabel('Age Histogram')
axs2[1].set_xlabel('Political Identity')
plt.show()
```

```python
fig3, ax3 = plt.subplots()
ax3.hist(cces['Evaluation'])

plt.show
```

```python
fig4, ax4 = plt.subplots()
ax4.scatter(cces['ConMediaUse'], cces['ImmiPhobia'])

plt.show
```

To improve this scatterplot, we could use different shapes, colors, or sizes of the dots for each of the variables on the scatterplot.  This information was found in the Directory of Visualization section 5.4 as well as in Visualizaing Data: Mapping Data onto Aesthetics shown in figure 2.5.
