```python
import plotly.express as px
import plotly as pl
import plotly.io as pio
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
npa = pd.read_csv('Nutrition__Physical_Activity__and_Obesity.csv')
npa.head()
```

```python
#Look for obesity info
npa = npa[npa["Topic"].str.contains("Obesity", case=False, na=False)]

#Get the correct value types
npa["Data_Value"] = pd.to_numeric(npa["Data_Value"], errors="coerce")

#plot information for specific info
plot_npa = npa.dropna(subset=["Age_yrs", "Data_Value", "LocationDesc"])
```

```python
npa.head()
```

```python
plot_npa = (
    plot_npa
    .groupby(["LocationDesc", "Age_yrs"])["Data_Value"]
    .mean()
    .reset_index()
)
```

```python
import plotly.graph_objects as go

states = plot_npa["LocationDesc"].unique()

fig = go.Figure()

# Add one trace per state
for state in states:
    state_npa = plot_npa[plot_npa["LocationDesc"] == state]
    
    fig.add_trace(
        go.Scatter(
            x=state_npa["Age_yrs"],
            y=state_npa["Data_Value"],
            mode="lines+markers",
            name=state,
            visible=(state == states[0]))
    )
fig.show()
```

```python
import plotly.graph_objects as go

states = plot_npa["LocationDesc"].unique()

fig1 = go.Figure()

# Add one trace per state
for state in states:
    state_npa = plot_npa[plot_npa["LocationDesc"] == state]
    
    fig1.add_trace(
        go.Scatter(
            x=state_npa["Age_yrs"],
            y=state_npa["Data_Value"],
            mode="lines+markers",
            name=state,
            visible=(state == states[0])
            )
    )

# Create dropdown buttons
buttons = []

for i, state in enumerate(states):
    visibility = [False] * len(states)
    visibility[i] = True
    
    buttons.append(
        dict(
            label=state,
            method="update",
            args=[
                {"visible": visibility},
                {"title": f"Obesity Rates by Age in {state}"}
            ]
        )
    )
# Add dropdown to layout
fig1.update_layout(
    title=f"Obesity Rates by Age in {states[0]}",
    xaxis_title="Age Group",
    yaxis_title="Obesity Rate (%)",
    updatemenus=[
        dict(
            active=0,
            buttons=buttons,
            x=1.15,
            y=1
        )
    ]
)

fig1.show()
```
