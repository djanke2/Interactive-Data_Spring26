```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
#New Dataset retrieved from https://catalog.data.gov/dataset/nutrition-physical-activity-and-obesity-behavioral-risk-factor-surveillance-system
npa = pd.read_csv('Nutrition__Physical_Activity__and_Obesity.csv')
npa.head()
```

```python
print(npa.columns)
```

```python
#Make columns numeric for ease of use later
npa["Data_Value"] = pd.to_numeric(npa["Data_Value"], errors="coerce")
npa["Low_Confidence_Limit"] = pd.to_numeric(npa["Low_Confidence_Limit"], errors="coerce")
npa["High_Confidence_Limit "] = pd.to_numeric(npa["High_Confidence_Limit "], errors="coerce")
npa["Sample_Size"] = pd.to_numeric(npa["Sample_Size"], errors="coerce")
```

```python
npa.head()
```

```python
#Make an overall Year column for later
npa["Year"] = npa["YearStart"]
```

```python
npa.head()
```

```python
#Records by gender pie chart
gender = npa["Sex"].value_counts().reset_index()
gender.columns = ["Sex", "Count"]
fig12 = px.pie(
    gender,
    names="Sex",
    values="Count",
    title="Distribution of Records by Gender")
fig12.show()
#This doesn't seem very telling as the info results in 50/50 - missing data needs to be cleaned up
```

```python
#Records by race/ethnicity pie cart
race_counts = npa["Race/Ethnicity"].value_counts().reset_index()
race_counts.columns = ["Race/Ethnicity", "Count"]
fig13 = px.pie(
    race_counts,
    names="Race/Ethnicity",
    values="Count",
    title="Distribution of Records by Race/Ethnicity")
fig13.show()
#This pie chart appears to tell more but all of the percentages are the same - missing data needs to be cleaned up.
```

```python
#Pie chart 
class_counts = npa["Class"].value_counts().reset_index()
class_counts.columns = ["Class", "Count"]
fig14 = px.pie(
    class_counts,
    names="Class",
    values="Count",
    title="Distribution by Health Classification")
fig14.show()
#Needs to be reviewed - data needs cleaned and what is this really telling the audience?
```

```python
#Averages by gender - Bar Chart

avg_sex = npa.groupby("Sex")["Data_Value"].mean().reset_index()
fig15 = px.bar(
    avg_sex,
    x="Sex",
    y="Data_Value",
    color="Sex",
    title="Average Data Value by Sex")
fig15.show()
```

```python
#Averages by race/ethnicity - Bar Chart
avg_race = npa.groupby("Race/Ethnicity")["Data_Value"].mean().reset_index()
fig16 = px.bar(
    avg_race,
    x="Race/Ethnicity",
    y="Data_Value",
    color="Race/Ethnicity",
    title="Average Data Value by Race/Ethnicity")
fig16.show()
```

```python
#Averages by top 10 states - Bar Chart
state_avg = npa.groupby("LocationDesc")["Data_Value"].mean().reset_index()
top_states = state_avg.sort_values(by="Data_Value", ascending=False).head(10)
fig17 = px.bar(
    top_states,
    x="LocationDesc",
    y="Data_Value",
    color="LocationDesc",
    title="Top 10 States by Average Data Value")
fig17.show()
```

```python
#Averages by Year and Gender - Bar Chart
year_sex = npa.groupby(["Year", "Sex"])["Data_Value"].mean().reset_index()
fig18 = px.bar(
    year_sex,
    x="Year",
    y="Data_Value",
    color="Sex",
    barmode="group",
    title="Average Data Value by Year and Gender")
fig18.show()
```

```python
#Scatterplot of data values and sample size
fig19 = px.scatter(
    npa,
    x="Sample_Size",
    y="Data_Value",
    color="Sex",
    hover_data=["LocationDesc", "Year"],
    title="Sample Size vs Data Value")
fig19.show()
```

```python
#Scatter Matrix of Data Value, Sample Size, & Confidence Limits
fig20 = px.scatter_matrix(
    npa,
    dimensions=["Data_Value", "Sample_Size", "Low_Confidence_Limit", "High_Confidence_Limit "],
    color="Sex",
    title="Scatter Matrix of Key Variables")
fig20.show()
```

```python
#Line Graph of Trend
trend = npa.groupby("Year")["Data_Value"].mean().reset_index()
fig21 = px.line(
    trend,
    x="Year",
    y="Data_Value",
    title="Average Data Value Over Time")
fig21.show()
```

```python
#Distribution Box plot. Maybe can help show variability?
fig22 = px.box(
    npa,
    x="Sex",
    y="Data_Value",
    color="Sex",
    title="Distribution of Data Values by Sex")
fig22.show()
```
